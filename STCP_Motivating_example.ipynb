```
# function create to get all N subsets
def sub_lists (l): 
    base = []   
    lists = [base] 
    for i in range(len(l)): 
        orig = lists[:] 
        new = l[i] 
        for j in range(len(lists)): 
            lists[j] = lists[j] + [new] 
        lists = orig + lists 
          
    return lists

subsets =  sub_lists(N)
#subsets.remove([])
```

R(S) - given a configuration ğ‘†âŠ‚ğ‘ and test reading R,
We define function **RS(S,R)** to return the set of all its signatures.



```
def RS(S,R):
  R_S = []
  for j in R:
    r = []
    for i in S:
        r.append(j[i-1])
    if r not in R_S:
      R_S.append(r)
  return (R_S)

# R(S) for ğ‘†={1,2}
RS((1,2), TR)
```

Q(S,r) - given a configuration ğ‘†âŠ‚ğ‘ and signature r,  We defined **function QSr(S,r)**  to return the set of all the possible readings from which signature ğ« can be obtained when the configuration is ğ‘†. ğ‘„(ğ‘†,ğ«)={ğ«Ìƒâˆˆğ‘…:ğ‘ŸÌƒğ‘–=ğ‘Ÿğ‘– âˆ€ğ‘–âˆˆğ‘†}.



```
def QSr(S,r):
  Q_S = []
  for j in TR:
    rj = []
    for i in S:
      rj.append(j[i-1])
    if rj == r :
      Q_S.append(j)
  return (Q_S)

# Q(S,r) for ğ‘†={1,2} and r = (0,0)
#QSr([1,2], [0,1])
```

**Formulation (1)** ğ‘(ğ‘˜)= Î£ğ‘(ğ‘˜|ğ«Ìƒ)ğ‘(ğ«Ìƒ) ; ğ«Ìƒâˆˆğ‘… - the prior probability of each class.

```
def Pk(k):
  total_prob = 0 
  for index_r_, r_ in enumerate(pr_):
    for index_kr_, kr_ in enumerate(pkr_[k]):
      if index_r_ == index_kr_:
        total_prob += r_ * kr_
  return (total_prob)

# Pk(k) for k = "Positive"
#Pk("Positive")
```

**Formulation (3)** Ps(ğ«|ğ‘˜)=Î£P(ğ«Ìƒ|ğ‘˜) ;
ğ«Ìƒâˆˆğ‘„(ğ‘†,ğ«) - the probability of a signature given class ğ‘˜.







```
# We created nested dictionary : first key is a class , second key is a reading, value is the conditional probability of the reading given the class.
# for example: {"Negative": {"(0,0,0)": 0.035 }}
dict_k_r_prk = {}
for key in pr_k:
  dict_k_r_prk.update({key: {}})
  for index_i, i in enumerate(TR):
    dict_k_r_prk[key].update({i:{}})
    for index_j, j in enumerate(pr_k[key]):
      if index_j == index_i:
        dict_k_r_prk[key][i] = j 
```

```
# Formulation (3) implementation
def PSrk(S,r,k):
  total_prob = 0
  QS = QSr(S,r)
  for r_ in QS:
    total_prob += dict_k_r_prk[k][r_]
  return (total_prob)

# PSrk(S,r,k) for ğ‘†={1,2} and r = (1,0) and k="Positive"
#PSrk((1,2,3),(1,0,0),"Positive")
```

**Formulation (4)** Ps(ğ«)=Î£P(ğ«Ìƒ) ; ğ‘ŸÌƒâˆˆğ‘„(ğ‘†,ğ«) - the prior probability of signature ğ«âˆˆğ‘…(ğ‘†).

```
# We created a dictionary : key is a reading, value is the a-priory probability of obtaining the reading ğ«Ìƒâˆˆğ‘….
# for example: {"(0,0,0)": 0.067 }
dict_r_pr_ = {}
for index_i, i in enumerate(TR):
    dict_r_pr_.update({i:{}})
    for index_j, j in enumerate(pr_ ):
      if index_j == index_i:
       dict_r_pr_[i] = j 
```

```
# Formulation (4) implementation
def PSr(S,r):
  total_prob = 0
  QS = QSr(S,r)
  for r_ in QS:
    total_prob += dict_r_pr_ [r_]
  return (total_prob)

# PSr(S,r) for ğ‘†={1,2} and r = (0,0)
#PSr([1],[1])
```

**Formulation (5)** Ps(ğ‘˜|ğ«)=Ps(ğ«|ğ‘˜)âˆ™ğ‘(ğ‘˜) / Ps(ğ«) - the a-posteriori probability that the class is ğ‘˜âˆˆğ¾ given that the signature is ğ«âˆˆğ‘…(ğ‘†).

```
def PSkr(S,r,k):
  return PSrk(S,r,k) * Pk(k) / PSr(S,r)

# PSkr(S,k,r) for ğ‘†={1,2} and r = (0,0) and k="Positive"
#PSkr((1,2),(0,0),"Positive")
```

**Formulation (6)** ğ¸s(ğ«|ğ‘™)= Î£ğœ†ğ‘˜ğ‘™âˆ™Ps(ğ‘˜|ğ«) ; ğ‘˜âˆˆğ¾ - Consider a given configuration ğ‘†âŠ‚ğ‘. The expected classification error cost for signature ğ« of ğ‘† if it is mapped to class ğ‘™

```
def ESrl(S,r,l):
  sum = 0 
  for k in K:
        sum += ğœ†ğ‘˜ğ‘™[K[k]][K[l]] * PSkr(S,r,k)
  return sum
  
# ESrl(S,r,l) for ğ‘†={1,2} and r = (0,0) and l="Positive"
#ESrl((1,2),(0,0),"Negative")
```

```
def lS(S,r):
  min = 100
  for l in K:
          if (ESrl(S,r,l) < min):
            min = ESrl(S,r,l)
            k = l
  return k

# lS(r) for r = (0,0,0)
#lS((1,2),(0,0))
```

```
def ES(S,r):
  min = 100
  for l in K:
          if (ESrl(S,r,l) < min):
            min = ESrl(S,r,l)
  return min

# ES(r) for r = (0,0,0)
#ES((1,2),(0,0)
#ES([],[])
```

**Formulation (9)** minğ‘†âŠ†ğ‘{Î£ğ‘ğ‘†(ğ«)ğ¸ğ‘†âˆ—(ğ«) ; ğ«âˆˆğ‘…(ğ‘†) +Î£ğ‘ğ‘– ; ğ‘–âˆˆğ‘†} - The STCP formulated mathematically

```
def STCP():
  min_total_error = 100
  for S in subsets:
    test_cost = 0 
    classification_error = 0 
    for r in RS(S,TR):
      classification_error += PSr(S,r) * ES(S,r)
    for i in S:
      test_cost += ci[i-1]
      total_error = classification_error + test_cost
      if (min_total_error> total_error):
        min_total_error = total_error
        min_s = S
  print(total_error)
  print(min_s)


STCP()
```
